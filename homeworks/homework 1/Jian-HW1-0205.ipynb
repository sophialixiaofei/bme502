```python
epsmax=1e-15   #define the maximum epsilon
epsmini=1e-20   #define the minimum epsilon
eps=(epsmax + epsmini)/2

while True:   #approach the machine epsilon by narrowing the span 
    if (1.0+eps)>1.0:
        epsmax=eps
        eps=(epsmax+epsmini)/2
        print eps
        print "still adding"

    else:
        epsmini=eps
        eps=(epsmax+epsmini)/2
        output=eps
        print eps
        print "too small"
        
        if output==epsmini:
            print "machine epsilon is %.50e." %eps   #output 50 decimals 
            break
```
